Write a program to plot
\begin{equation}
f(x)=\pi/2+2*(sin(x)+sin(3x)/3+sin(5x)/5+....sin(nx)/n)
\end{equation}
from -$\pi$ to $\pi$. Choose n=11, 21 and 31 and plot all these three curves in one figure. Please use numpy library and matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-np.pi,np.pi+.01,.01)

def f(stp,x): #This section lets me create a for loop function that I can call instead of recoding everything
    func=np.pi/2
    n=np.arange(1,stp+1,2) #stp+1 lets is just so the last term of the sequence is equal to stp
    for i in n:
        func+=2*np.sin(i*x)/i #here I add each consective term in the series
    return func

plt.figure(figsize=(10,8)) #adjusted figure size to be easier to see

plt.plot(x,f(11,x),label='n=11')

plt.plot(x,f(21,x),label='n=21')

plt.plot(x,f(31,x),label='n=31')

plt.legend() #created three different plot functions and the legend just adds label from each function

plt.show()
```
