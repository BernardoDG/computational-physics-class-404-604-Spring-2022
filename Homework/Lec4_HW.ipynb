# Part 1
Evaluate the derivative at position x using the forward difference scheme. Plot log(Error)=log10|f‚Äô(numerically)-f‚Äô(analytic)| against log10(dx) for dx=1e-16,1e-15,‚Ä¶.,0.1 at x=œÄ/4. You should find Error~ùëëùë•ùëõ; what is n? what is the optimal value for dx.

```python
import numpy as np
import matplotlib.pyplot as plt

def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx

x=np.pi/4
dxarr=[]
err=[]
for dx in 10**np.arange(-16, -.8, .2):
    error=np.abs(derifw(np.cos,x,dx)+np.sin(x))
    dxarr.append(dx)
    err.append(error)
    
plt.plot(np.log10(dxarr), np.log10(err), 'b-', lw=2)     
plt.xlabel('dx')
plt.ylabel('Error')
plt.show()
#Log-Log polynomials become linear so the order of the error can be calculated by the slope of the line after
# the rounding error becomes negligable
print((np.log10(err[-1])-np.log10(err[-2]))/(np.log10(dxarr[-1])-np.log10(dxarr[-2])))
min_err=min(err)
print(dxarr[err.index(min_err)])
```

From the slope of the line we can see that n is approximately 1. The optimum value of dx is 1.58e-8.

```python
def dericd(y,x,dx):
    return (y(x+dx/2)-y(x-dx/2))/(dx)

x=np.pi/4
dxarr=[]
err=[]
for dx in 10**np.arange(-16., -0.8, 0.2):
    error=np.abs(dericd(np.cos,x,dx)+np.sin(x))
    dxarr.append(dx)
    err.append(error)
    
plt.plot(np.log10(dxarr), np.log10(err), 'b-', lw=2)    

plt.show()

print((np.log10(err[-1])-np.log10(err[-2]))/(np.log10(dxarr[-1])-np.log10(dxarr[-2])))
min_err=min(err)
print(dxarr[err.index(min_err)])
```

The value of n is approximately 2 and the optimum value of dx is 1e-5. Since error with this scheme was smaller that the foward difference but achived with a larger value of dx, this this scheme is better.
